This notebook serves as a testing environment for TCMIX's namelist parameters that define initial temperature/salinity profiles.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Profile functions ---
def temp_profile(z, t_surf=28.0, t_deep=4.0, z_mid=100.0, dz_tcl=40.0):
    norm = 1 - np.tanh(-z_mid / dz_tcl)
    return t_deep + (t_surf - t_deep) * (1 - np.tanh((z - z_mid) / dz_tcl)) / norm

def sal_profile(z, s_surf=36, s_deep=36.5, z_mid=180.0, dz_scl=80.0):
    norm = 1 - np.tanh(-z_mid / dz_scl)
    return s_deep + (s_surf - s_deep) * (1 - np.tanh((z - z_mid) / dz_scl)) / norm

# --- Depth vector (0 = surface, 1000m = deep ocean) ---
depth = np.linspace(0, 1000, 500)

# --- Compute profiles ---
temperature = temp_profile(depth)
salinity = sal_profile(depth)

# --- Plot ---
fig, ax1 = plt.subplots(figsize=(6, 8))

ax1.plot(temperature, depth, 'r-', label='Temperature (°C)')
ax1.set_xlabel('Temperature (°C)', color='r')
ax1.set_ylabel('Depth (m)')
ax1.invert_yaxis()
ax1.tick_params(axis='x', colors='r')

# Create second x-axis for salinity
ax2 = ax1.twiny()
ax2.plot(salinity, depth, 'b', label='Salinity (PSU)')
ax2.set_xlabel('Salinity (PSU)', color='b')
ax2.tick_params(axis='x', colors='b')

plt.title('Idealized Ocean Profiles for Tropical Cyclone Forcing')
plt.grid(True)
plt.tight_layout()
plt.show()
```
